# Factorials and Binomial Coefficients

```r
print(factorial(x = 5))
print(choose(n = 5, k = 2))

# The functions below compute log of n! and log of n-choose-k
# This functionality is given as beyond a point factorial(n) = infinity
print(lfactorial(x = 5))
print(lchoose(n = 5, k = 2))
```

# Sampling and Simulation

```r
print(sample(x = 10, size = 5))
print(sample(x = 10, size = 5, replace = TRUE))
print(sample(x = 10))
print(sample(x = letters, size = 7))
print(sample(x = 4, size = 3, replace = TRUE, prob = c(0.1, 0.2, 0.3, 0.4)))
```

# Matching Problem Simulation

```r
n <- 100 # number of cards in the deck

# sum(sample(n)==(1:n)) : In one experiment, see how many sampled cards are of the same number as their sampling position
# replicate this experiment 10^4 times
r <- replicate(n = 10^4, expr = sum(sample(n)==(1:n)))

# If for an experiment, there is atleast one card that has the same number as the sampling position, the corresponding 
# entry in vector 'r' will be >= 1. Find all such entries (favourable outcomes) and divide by the total number of outcomes.
sum(r>=1)/10^4
```
